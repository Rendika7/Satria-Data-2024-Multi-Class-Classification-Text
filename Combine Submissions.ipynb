# Combine

```python
# Menggabungkan data berdasarkan IDText
merged_data = pd.merge(sub_45, sub_62, on='IDText', suffixes=('_sub_45', '_sub_62'))

# Inisialisasi list untuk menyimpan hasil zip
zipped_data = []

# Looping untuk bandingkan kelas dan hasilkan zip jika kelas sama
for index, row in merged_data.iterrows():
    if row['Kelas_sub_45'] == row['Kelas_sub_62']:
        zipped_data.append((row['IDText'], row['Kelas_sub_45']))
    else:
        zipped_data.append((row['IDText'], False))

# Buat DataFrame baru dari zipped_data
zipped_df = pd.DataFrame(zipped_data, columns=['IDText', 'Kelas_Zip'])
```

```python
zipped_df.to_excel("zipped_df.xlsx", index=False)
```

```python
id_texts = (zipped_df[zipped_df["Kelas_Zip"] == False]["IDText"]).tolist()
```

```python
filtered_sub_45 = sub_45[sub_45['IDText'].isin(id_texts)]
filtered_sub_62 = sub_62[sub_62['IDText'].isin(id_texts)]
```
